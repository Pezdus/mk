```python
import matplotlib.pyplot as plt
import numpy as np

# Definicja funkcji sigmoid
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Definicja gradientu (pochodnej) funkcji sigmoid
def sigmoid_gradient(x):
    s = sigmoid(x)
    return s * (1 - s)

# 100 punktów od -5 do 5
x = np.linspace(-5, 5, 100)

# Obliczenie wartości funkcji oraz gradientu
y = sigmoid(x)
dy = sigmoid_gradient(x)

# Tworzenie wykresu
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(1, 1, 1)

# Ustawienie osi w centrum
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# Rysowanie funkcji Sigmoid (linia czerwona) i jej gradientu (linia niebieska, przerywana)
plt.plot(x, y, 'r', label='Sigmoid: $f(x)$', linewidth=2)
plt.plot(x, dy, 'b--', label="Gradient: $f'(x)$", linewidth=2)

# Dodanie siatki i legendy
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend(loc='upper left')
plt.title('Funkcja Sigmoid i jej gradient (pochodna)', pad=20)

# Pokazanie wykresu
plt.show()
```
