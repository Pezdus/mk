```python
import pandas as pd
import numpy as np

df = pd.read_csv('war7.csv', sep=';', decimal=',')

print("Wczytane dane:")
print(df)
print("-" * 30)

X = df[['x1', 'x2']].values
y = df['y'].values

X_pinv = np.linalg.pinv(X)

beta = X_pinv @ y

a = beta[0]
b = beta[1]

print("Wyniki:")
print(f"Współczynnik a: {a:.8f}")
print(f"Współczynnik b: {b:.8f}")


y_pred = X @ beta
print("-" * 30)
print("Rzeczywiste y vs Przewidywane y (dla sprawdzenia):")
for real, pred in zip(y, y_pred):
    print(f"Rzeczywiste: {real:.8f} | Przewidywane: {pred:.8f}")
```
